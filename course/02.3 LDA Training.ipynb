# Training our LDA Model

First, we need to load the result from the previous step:

```python
import json

with open('../data/articles/train.preprocessed.json', 'r') as f:
	data = json.load(f)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
vectorizer = CountVectorizer(min_df=0.01, max_df=0.85)
vectorizer.get_feature_names = vectorizer.get_feature_names_out

tfidf = TfidfTransformer()

X = vectorizer.fit_transform([ " ".join(doc) for doc in data ])
X = tfidf.fit_transform(X)

N, V = X.shape
print("Found a vocabulary of {} unique terms.".format(V))
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda = LatentDirichletAllocation(n_components=10, topic_word_prior=None, doc_topic_prior=None)
Y = lda.fit_transform(X)
```

```python
import pyLDAvis.sklearn
import pyLDAvis
pyLDAvis.enable_notebook()
```

```python
pyLDAvis.sklearn.prepare(lda, X, vectorizer, mds='tsne')
```
